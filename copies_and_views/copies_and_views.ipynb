```python
# Python default lists shares memory
x = list(range(1,10))
x
```

```python
y = x
y
```

```python
x[0] = "Apple"
x
```

```python
y # x and y are sharing memory
```

```python
# we can use list() to create a new copy
z = list(x)
z
```

```python
x[0] = "Orange"
x
```

```python
z # Still z[0] is "Apple" since x and z dont share memory
```
