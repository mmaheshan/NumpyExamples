```python
import numpy as np
x = np.arange(1,101)
x
```

```python
y = x
x[0] = 10000
y # While values of x[0] was changed, y also changed
```

```python
y = x[1:10]
y
```

```python
# Now I change x, y still change
x[1] = 2000
y
```

```python
# How to know if variables are sharing memory?
np.may_share_memory(x, y)
```

```python
# What if is I did this:
y = x[1:10].astype(float)
y
```

```python
# They dont share memory now!
np.may_share_memory(x , y)
```

```python
x = np.arange(1,101).reshape(10,10)
x
```

```python
y = x[:]
y.shape
```

```python
x.dtype
```

```python
np.may_share_memory(x , y)
```

```python
y = x.reshape(5,20).T
y
```

```python
np.may_share_memory(x, y)
```

```python
# Copy operation prevents memory sharing
y = x.reshape(5,20).T.copy() 
y
```

```python
np.may_share_memory(x,y)
```
