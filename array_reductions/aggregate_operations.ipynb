```python
# Create a 1D array of numbers from 1 to 10
row_template = np.arange(1, 11)

# Repeat this row 10 times to create a 2D array
array_1_to_10_repeated = np.tile(row_template, (10, 1))

print(array_1_to_10_repeated)
```

```python
np.sum(array_1_to_10_repeated, axis=0) # Add all columnwise 
```

```python
np.sum(array_1_to_10_repeated, axis=1) # Add all rowwise
```

```python
array_1_to_10_repeated.sum()
```

```python
array_1_to_10_repeated.mean()
```
