# Operations on different data types

```python
x = np.arange(0.0, 10.0, 1.0) # creates a float array
x
```

```python
np.info(x)
```

```python
y = np.arange(10) # created a integer array
y
```

```python
np.info(y)
```

```python
x + y # We get a float array. Basically integer values have been upcasted to float
```

```python
y.dtype # y is integer array
```

```python
# Assigning a floar into a integer array
y[0] = 13.8
y.dtype # Data Type of y stayed same as integer !!
```

```python
# are you sure?!!
y
```

```python
# The assigend value has been downcasted to int
```
