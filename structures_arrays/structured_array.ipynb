# Structured Arrays


```python
import numpy as np

letters =list('abcdefghijklmnopqrstuvwxyz')
numbers = list(range(1,27))
letters, numbers
```

```python
np_letters = np.array(letters)
np_numbers = np.array(numbers)
np_letters, np_numbers
```

```python
np_letters[np_numbers == 10]
```

```python
#ciphertype = np.dtype([('letter','S20'), ('number',int)])
ciphertype = np.dtype([('letter','>S20'), ('number','>i8')])
```

```python
data = list(zip(letters, numbers))
data
```

```python
cipher = np.array(data, dtype=ciphertype)
cipher
```

```python
cipher["number"] + cipher["number"] # Add a column from data type 
```

## Sorting a data structure

```python
data = list(zip(letters, reversed(numbers)))
data
```

```python
cipher = np.array(data, dtype=ciphertype)
cipher
```

```python
np.sort(cipher, order="number") # Sort a data structure
```

```python
# 1. Define the custom structured data type
# This dtype represents a 'Person' with fields for 'name', 'age', and 'height'.
# 'name' is a string of up to 20 characters ('S20'), 'age' is an unsigned 8-bit integer ('u1'),
# and 'height' is a 32-bit floating-point number ('f4').
person_dtype = np.dtype([
    ('name', 'S20'),
    ('age', 'u1'),
    ('height', 'f4')
])

# 2. Create a NumPy array using the custom dtype
# The array is initialized with a list of tuples, where each tuple corresponds
# to a record in the structured array.
people_data = np.array([
    ("Alice", 30, 1.65),
    ("Bob", 25, 1.80),
    ("Charlie", 35, 1.75)
], dtype=person_dtype)

# 3. Accessing data by field name
print("Accessing specific fields:")
print(f"Names: {people_data['name']}")
print(f"Ages: {people_data['age']}")
print(f"Heights: {people_data['height']}")

# 4. Accessing individual records
print("\nAccessing individual records:")
print(f"First person: {people_data[0]}")
print(f"Second person's age: {people_data[1]['age']}")

# 5. Modifying data
print("\nModifying data:")
people_data[0]['age'] = 31
print(f"Alice's updated age: {people_data[0]['age']}")
```
