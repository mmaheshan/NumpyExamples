# Array Operations like shape, ravel(), flatten(), reshape(), resize()

```python
import numpy as np

x = np.arange(100)
x, x.shape
```

```python
y = np.arange(10)
y
```

```python
z = np.random.rand(10,10)
z, z.shape
```

```python
k = z.ravel() # Converts the shape of a multi dimensional arrays to a flat shape
k , k.shape
```

```python
z[:3,:3], z[:3,:3].ravel(), z[:3,:3].flatten()
```

```python
z = np.random.rand(3,3)
z
```

```python
a = z[:3,:3].ravel()
b = z[:3,:3].flatten()
```

```python
np.may_share_memory(z, a) # Ravel shares memory!
```

```python
np.may_share_memory(z, b) # Flatten do not share memory!
```

```python
arr = np.ndarray(100) # Same as np.random.rand() -> which give ndadarray as output
arr
```

```python
arr1 = arr.reshape(10,10) # Simple reshape
arr1
```

```python
arr1 = arr.reshape(-1, 20) # Leave one dimension for numpy calcualtion
arr1.shape
```

```python
arr1 = arr.reshape(4,5,-1)
arr1.shape
```

```python
y = np.arange(10)
y, y.shape
```

```python
k = np.resize(y, 20)
k , k.shape, y.shape
```

```python
y1 = np.arange(10)
print(y1)
y1.resize(20) # There is a difference in np.resize and ndarray.resize
# np.resize fills additionally added cells with same data 
# ndarray.resize fills with 0
y1, y1.shape
```

```python
# With string array it works while filling with blanks
l = np.array(['a', 'b','c'])
print(l)
l.resize(10)
l
```

```python
# Boolean fills with False
q = np.array([True, True, True])
print(q)
q.resize(10)
q
```
