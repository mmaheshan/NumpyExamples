# Time Series Analysis

```python
import numpy as np
import csv
import matplotlib.pyplot as plt

with open("heart-rate-time-series.csv") as f:
    reader = csv.reader(f)
    raw_data = list(reader)


raw_data
```

```python
np.array(raw_data).shape, np.array(raw_data).dtype
```

```python
np.array(raw_data).squeeze().shape
```

```python
heartrate= np.array(raw_data).squeeze().astype(np.float32)
heartrate
```

```python
time = np.linspace(0, 900, 1800)
time
```

```python
fig, ax = plt.subplots(figsize=(20, 20))
ax.plot(time, heartrate)
ax.plot([time.min(), time.max()],[heartrate.mean(), heartrate.mean()],'red')
```

```python
query = time < 200
query
```

```python
fig, ax = plt.subplots(figsize=(20,10))
ax.plot(time[query], heartrate[query])
```

```python
def describe(arr):
    return dict(mean = arr.mean(),
                min = arr.min(),
                max = arr.max(),
                median = np.median(arr),
                std = arr.std())

describe(heartrate)
```
